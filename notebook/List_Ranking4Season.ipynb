```python
from typing import Dict, List
import pandas as pd
import re


def make_ranking(matches: Dict[str, Dict[str, pd.DataFrame]]) -> List[Dict[str, str]]:
    """与えられた試合データから、勝ち点、得失点などを元に順位リス尾を作成
    """
    _res = [{'name': _team,
             'point': matches[_team]['df']['point'].astype('int').sum(),
             'goal_diff': matches[_team]['df']['goal_get'].astype('int').sum() - matches[_team]['df']['goal_lose'].astype('int').sum(),
             'goal_get': matches[_team]['df']['goal_get'].astype('int').sum()
            } for _team in matches]
    return [clb for clb in sorted(_res, key=lambda x:(x['point'], x['goal_diff'], x['goal_get']), reverse=True)]

_tmp = pd.DataFrame()
for _file in reversed(glob('../csv/*-J?.csv')):
    season = re.match(r'\.\./csv/(\d+[AB]?)', _file)[1]
    _df = read_allmatches_csv(_file)
    _map = make_team_map(_df)
    ranking = [x['name'] for x in make_ranking(_map)]
    print('["' + '", "'.join(ranking) + '"]')
    if season.endswith('B'):
        _tmp = _df
    elif season.endswith('A'):
        _df = pd.concat([_tmp, _df])
        ranking = [x['name'] for x in make_ranking(make_team_map(_df))]
        print(season[:-1])
        print('["' + '", "'.join(ranking) + '"]')
        _tmp = pd.DataFrame()
```
